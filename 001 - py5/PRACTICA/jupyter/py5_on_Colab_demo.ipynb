## ¡Ahora a probar **py5**!

```python
import py5
import py5_tools
# Carga py5 y sus herramientas auxiliares para ejecutar y capturar el sketch.

```

```python
def setup():
  py5.size(600, 600)
  for _ in range(100):
      py5.fill(py5.random_choice(('red', 'green', 'blue')), 100)
      py5.square(py5.random(50, 500), py5.random(50, 500), py5.random(10, 30))
# Define la función setup del sketch: crea un lienzo y dibuja cuadrados aleatorios de colores.
```

```python
py5.run_sketch()
# Inicia la ejecución del sketch definido en setup().
```

A diferencia de una versión local del cuaderno, ejecutándose en Jupyter Lab, por ejemplo, no verás una ventana con tu sketch ejecutándose.

Pero sí está corriendo y puedes tomar una captura usando  `py5_tools.screenshot()`. Después puedes detenerlo usando `py5.exit_sketch()`

```python
py5_tools.screenshot()
# Toma una captura del sketch, útil cuando no hay una ventana visible como en Colab.
```

WOW!

```python
# If you want to run it again, stop the sketch.
py5.exit_sketch()
# Detiene el sketch para poder volver a ejecutarlo sin reiniciar todo el notebook.
```
