```python
# Importo librerias del sistema, manejo de rutas y descarga de archivos:
import sys # sys: da informacion del entorno de Python y del sistema
import urllib.request # urllib.request: permite descargar archivos desde una URL
from pathlib import Path # Path: facilita trabajar con rutas y archivos de forma mas clara

# Importo librerias para procesamiento y visualizacion de imagenes:
import cv2 # cv2: OpenCV, usada para leer, procesar y transformar imagenes
import matplotlib.pyplot as plt # matplotlib.pyplot: muestra imagenes y graficos dentro del notebook
import numpy as np # numpy: trabaja con matrices y arrays, base para representar imagenes
import skimage as ski # skimage: scikit-image, aporta funciones de analisis y procesamiento de imagenes
from PIL import Image # Image: clase de Pillow para abrir, convertir y guardar imagenes

print(f"Versión de Python: {sys.version.split()[0]}")
print(f"NumPy: {np.__version__}")
print(f"OpenCV: {cv2.__version__}")
print(f"scikit-image: {ski.__version__}")
print(f"Matplotlib: {plt.matplotlib.__version__}")
```

## ¿Para qué sirve cada biblioteca?

- `NumPy` permite trabajar con matrices. En procesamiento de imágenes esto es clave, porque una imagen digital se representa como una matriz de números.
- `OpenCV` ofrece herramientas muy usadas en visión por computadora.
- `Matplotlib` nos ayuda a visualizar imágenes, gráficos e histogramas.
- `PIL` es útil para abrir, guardar y manipular imágenes desde otra perspectiva.
- `scikit-image` reúne funciones diseñadas específicamente para análisis y procesamiento de imágenes.

No hacen exactamente lo mismo. Por eso conviene conocer el papel de cada una.


```python
ruta_imagen = Path("frutos_rojos.png")
url = "https://www.dropbox.com/s/do6b49d10ty1dyf/frutos_rojos.png?dl=1"

if not ruta_imagen.exists():
    urllib.request.urlretrieve(url, ruta_imagen)

print(f"Imagen disponible en: {ruta_imagen.resolve()}")
# Define la ruta local de la imagen y la descarga desde internet solo si todavia no existe en la carpeta del proyecto.
# Al final muestra la ubicacion completa del archivo para verificar que quedo disponible.

```

```python
# Carga la misma imagen con OpenCV, Matplotlib y Pillow para comparar
# como cada libreria la representa, que informacion devuelve y como se visualiza.
# En OpenCV convierto de BGR a RGB para que los colores se vean correctamente al mostrarla.


img_cv = cv2.imread(str(ruta_imagen)) #carga la imagen con OpenCV
img_cv_rgb = cv2.cvtColor(img_cv, cv2.COLOR_BGR2RGB) #corrige el orden de color para poder mostrarla bien OpenCV suele leer en BGR, no en RGB
img_plt = plt.imread(ruta_imagen) #carga la imagen con Matplotlib
img_pil = Image.open(ruta_imagen) #la abre con Pillow

#ahora imprimo la información de las imágenes
print("OpenCV")
print(f"  forma: {img_cv.shape}")
print(f"  tipo de dato: {img_cv.dtype}")
print()
print("Matplotlib")
print(f"  forma: {img_plt.shape}")
print(f"  tipo de dato: {img_plt.dtype}")
print()
print("PIL")
print(f"  tamaño: {img_pil.size}")
print(f"  modo: {img_pil.mode}")

fig, axes = plt.subplots(1, 4, figsize=(15, 5))

axes[0].imshow(img_cv_rgb)
axes[0].set_title("OpenCV + conversión a RGB")
axes[0].axis("off")

axes[1].imshow(img_plt)
axes[1].set_title("Matplotlib")
axes[1].axis("off")

axes[2].imshow(img_pil)
axes[2].set_title("PIL")
axes[2].axis("off")

#imprimo una extra para ver la diferencia de la imagen original de OpenCV y la correccion realizada
axes[3].imshow(img_cv)
axes[3].set_title("OpenCV")
axes[3].axis("off")

plt.tight_layout()
plt.show()
```

## Para mirar con calma

- `OpenCV` devuelve un arreglo de `NumPy`.
- `Matplotlib` también trabaja con arreglos, pero no siempre usa el mismo tipo de dato que `OpenCV`.
- `PIL` abre la imagen como un objeto distinto, pensado para otras tareas.

Lo importante es que una imagen no entra siempre al entorno del mismo modo. Antes de procesarla, conviene revisar su forma, su tipo de dato y la biblioteca con la que fue cargada.
