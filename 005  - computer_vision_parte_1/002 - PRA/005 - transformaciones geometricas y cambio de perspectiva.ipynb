# Transformaciones geométricas y cambio de perspectiva

En este cuaderno vas a rectificar una carta fotografiada en perspectiva para verla de frente. La idea central es entender que la transformación depende de elegir bien puntos de origen y puntos de destino.


## Objetivo

Construir una transformación de perspectiva a partir de cuatro esquinas y usarla para obtener una versión frontal de una carta dentro de una fotografía.

## Resultados de aprendizaje

Al finalizar este cuaderno, vas a poder:

- identificar coordenadas de origen sobre una imagen;
- definir una geometría de destino ordenada;
- calcular una matriz de perspectiva con OpenCV;
- interpretar qué cambia y qué se preserva en una rectificación.

## Relación con la secuencia

Este cuaderno retoma la lectura de coordenadas trabajada en operaciones básicas y la lleva a una transformación más específica. Más adelante, la lógica de elegir puntos y transformar regiones vuelve a aparecer en tareas de anotación y segmentación.


## Módulos que vamos a usar

- `cv2`: para calcular la matriz de perspectiva y aplicar la transformación.
- `numpy`: para organizar las coordenadas en arreglos de puntos.
- `matplotlib.pyplot`: para comparar la imagen original, los puntos marcados y el resultado.
- `pathlib.Path`: para ubicar la imagen de trabajo.


```python
from pathlib import Path

import cv2
import numpy as np
import matplotlib.pyplot as plt

ruta_imagen = Path("Imagenes") / "cards.jpg"
imagen_bgr = cv2.imread(str(ruta_imagen), cv2.IMREAD_COLOR)
if imagen_bgr is None:
    raise FileNotFoundError(f"No se pudo leer la imagen: {ruta_imagen}")

imagen_rgb = cv2.cvtColor(imagen_bgr, cv2.COLOR_BGR2RGB)
```

## 1. Imagen original

Primero miramos la escena completa. Después vamos a seleccionar una sola carta y marcar sus cuatro esquinas.


```python
plt.figure(figsize=(8, 6), constrained_layout=True)
plt.imshow(imagen_rgb)
plt.title("Imagen original con cartas", fontweight="bold", loc="left")
plt.axis("off")
plt.show()
```

## 2. Elegir puntos de origen

Para corregir la perspectiva no alcanza con decir “quiero esa carta”. Necesitamos cuatro puntos bien ubicados. Esos puntos describen el cuadrilátero que vamos a transformar.


```python
# Define las cuatro esquinas de la carta seleccionada, las marca sobre una copia de la
# imagen original y muestra el resultado para verificar visualmente los puntos elegidos
# antes de aplicar la transformación de perspectiva.


# Define las coordenadas de las cuatro esquinas de la carta en la imagen original.
# Estos puntos se usarán como referencia de origen para corregir la perspectiva.

esquinas_origen = np.array(
    [
        [111, 219],
        [287, 188],
        [154, 482],
        [352, 440],
    ],
    dtype=np.float32,
)

# Crea una copia de la imagen y dibuja sobre ella las esquinas seleccionadas,
# etiquetándolas para verificar visualmente su ubicación y orden.

imagen_con_esquinas = imagen_rgb.copy()
for numero, punto in enumerate(esquinas_origen, start=1):
    x, y = int(punto[0]), int(punto[1])
    cv2.circle(imagen_con_esquinas, (x, y), 8, (255, 255, 255), -1)
    cv2.putText(
        imagen_con_esquinas,
        f"P{numero}",
        (x + 8, y - 8),
        cv2.FONT_HERSHEY_SIMPLEX,
        0.7,
        (255, 255, 255),
        2,
    )

plt.figure(figsize=(8, 6), constrained_layout=True)
plt.imshow(imagen_con_esquinas)
plt.title("Carta seleccionada con sus esquinas", fontweight="bold", loc="left")
plt.axis("off")
plt.show()
```

Fijate que el orden de los puntos importa. Si cambiás el orden, la transformación puede deformar la carta en vez de rectificarla. Por eso conviene mantener una lógica estable, por ejemplo: arriba-izquierda, arriba-derecha, abajo-izquierda, abajo-derecha.


### Guía práctica para seleccionar puntos en una imagen:
- El origen de coordenadas está en la esquina superior izquierda: (0, 0).# 
- El eje x crece hacia la derecha.
- El eje y crece hacia abajo.
- Entonces:
    * si te movés a la derecha, aumenta x
    * si te movés a la izquierda, disminuye x
    * si te movés hacia abajo, aumenta y
    * si te movés hacia arriba, disminuye y

 Referencias útiles:
 - Un punto se escribe como (x, y).
 - x representa la columna.
 - y representa la fila.
 - Si la imagen tiene shape (alto, ancho, canales):
    * x puede ir de 0 a ancho - 1
    * y puede ir de 0 a alto - 1

## 3. Definir los puntos de destino

Ahora vamos a decidir cómo queremos ver esa carta al final: como un rectángulo frontal con ancho y alto bien definidos.


```python
# Define el tamaño que tendrá la imagen rectificada después del cambio de perspectiva.
ancho_destino = 220
alto_destino = 320


# Define las cuatro esquinas del rectángulo de destino, que representan la forma
# idealizada donde se quiere proyectar la carta una vez corregida la perspectiva.
esquinas_destino = np.array(
    [
        [0, 0],
        [ancho_destino - 1, 0],
        [0, alto_destino - 1],
        [ancho_destino - 1, alto_destino - 1],
    ],
    dtype=np.float32,
)

# Crea un lienzo vacío y dibuja sobre él las esquinas de destino para visualizar
# la geometría final que se usará como referencia en la transformación.
lienzo_destino = np.zeros((alto_destino, ancho_destino, 3), dtype=np.uint8)
for numero, punto in enumerate(esquinas_destino, start=1):
    x, y = int(punto[0]), int(punto[1])
    cv2.circle(lienzo_destino, (x, y), 8, (255, 255, 255), -1)
    cv2.putText(lienzo_destino, f"D{numero}", (x + 8, min(y + 22, alto_destino - 10)), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 255), 2)

# Muestra la disposición de los puntos de destino para comprobar visualmente
# cómo quedará definida la salida rectificada.
plt.figure(figsize=(5, 6), constrained_layout=True)
plt.imshow(lienzo_destino)
plt.title("Geometría de destino", fontweight="bold", loc="left")
plt.axis("off")
plt.show()

```

## 4. Calcular la transformación y rectificar la carta

Con los puntos de origen y de destino ya definidos, OpenCV puede calcular una matriz que describe cómo pasar de una geometría a la otra.


```python
# Calcula la matriz de transformación de perspectiva a partir de las esquinas
# seleccionadas en la imagen original y las esquinas definidas en el destino.
matriz_perspectiva = cv2.getPerspectiveTransform(esquinas_origen, esquinas_destino)

# Aplica la transformación para proyectar la carta sobre la geometría de destino
# y obtener una versión rectificada con el tamaño especificado.
imagen_rectificada = cv2.warpPerspective(imagen_rgb, matriz_perspectiva, (ancho_destino, alto_destino))

# Compara la imagen original, los puntos de origen marcados y el resultado final
# de la rectificación para verificar visualmente la corrección de perspectiva.
fig, ejes = plt.subplots(1, 3, figsize=(15, 5), constrained_layout=True)
ejes[0].imshow(imagen_rgb)
ejes[0].set_title("Imagen original", fontweight="bold", loc="left")
ejes[0].axis("off")

ejes[1].imshow(imagen_con_esquinas)
ejes[1].set_title("Puntos de origen", fontweight="bold", loc="left")
ejes[1].axis("off")

ejes[2].imshow(imagen_rectificada)
ejes[2].set_title("Carta rectificada", fontweight="bold", loc="left")
ejes[2].axis("off")
plt.show()

```

El resultado no crea información nueva, pero reorganiza la que ya estaba en la fotografía. La carta sigue siendo la misma, aunque ahora se ve más cercana a una vista frontal.

```python
# Define manualmente las cuatro esquinas de otra carta en la imagen original.
# Ajustá estos valores cambiando x e y hasta que correspondan a la nueva carta elegida.
nuevas_esquinas_origen = np.array(
    [
        [280, 130],  # esquina superior izquierda
        [450, 130],  # esquina superior derecha
        [280, 350],  # esquina inferior izquierda
        [450, 350],  # esquina inferior derecha
    ],
    dtype=np.float32,
)

# Crea una copia de la imagen y marca sobre ella los nuevos puntos seleccionados
# para comprobar visualmente que correspondan a la carta que querés rectificar.
imagen_nueva_con_esquinas = imagen_rgb.copy()
for numero, punto in enumerate(nuevas_esquinas_origen, start=1):
    x, y = int(punto[0]), int(punto[1])
    cv2.circle(imagen_nueva_con_esquinas, (x, y), 8, (255, 255, 255), -1)
    cv2.putText(
        imagen_nueva_con_esquinas,
        f"P{numero}",
        (x + 8, y - 8),
        cv2.FONT_HERSHEY_SIMPLEX,
        0.7,
        (255, 255, 255),
        2,
    )

# Calcula la nueva transformación de perspectiva a partir de las esquinas elegidas
# y genera una segunda carta rectificada con la misma geometría de destino.
matriz_perspectiva_nueva = cv2.getPerspectiveTransform(nuevas_esquinas_origen, esquinas_destino)
nueva_carta_rectificada = cv2.warpPerspective(
    imagen_rgb,
    matriz_perspectiva_nueva,
    (ancho_destino, alto_destino),
)

# Muestra la imagen original, los nuevos puntos de origen y la carta rectificada
# para verificar si la selección manual produjo una corrección adecuada.
fig, ejes = plt.subplots(1, 3, figsize=(15, 5), constrained_layout=True)
ejes[0].imshow(imagen_rgb)
ejes[0].set_title("Imagen original", fontweight="bold", loc="left")
ejes[0].axis("off")

ejes[1].imshow(imagen_nueva_con_esquinas)
ejes[1].set_title("Nueva selección manual", fontweight="bold", loc="left")
ejes[1].axis("off")

ejes[2].imshow(nueva_carta_rectificada)
ejes[2].set_title("Nueva carta rectificada", fontweight="bold", loc="left")
ejes[2].axis("off")
plt.show()
```
